```python
import pandas as pd
from dotenv import load_dotenv
from pathlib import Path
import os 
import requests
import json
import time
```

```python
# Local & Constants
address_lookup = "5 Crieff Ave, Vaughan, ON L6A 2B6"
```

```python
# Get API Keys
# Load and export environment variables with load_dotenv()
path_key = Path(".env")
load_dotenv(dotenv_path=path_key)

# Retrieve environment variable and store in Python variable
api_key = os.getenv("RAPID_API_KEY")
```

```python
# Lookup for zpid for the property

url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"

querystring = {"location":address_lookup}

headers = {
    'x-rapidapi-host': "zillow-com1.p.rapidapi.com",
    'x-rapidapi-key': api_key
    }

zpid = requests.request("GET", url, headers=headers, params=querystring)

# Transfer to json
zpid = zpid.json()
zpid
```

```python
# Get full info. for the property from Zillow
url = "https://zillow-com1.p.rapidapi.com/property"

querystring = {"zpid":zpid["zpid"]}

headers = {
    'x-rapidapi-host': "zillow-com1.p.rapidapi.com",
    'x-rapidapi-key': api_key
    }

response = requests.request("GET", url, headers=headers, params=querystring)

response_content = response.content
print(response_content)
```

```python
# Format data as json
data = response.json()

print(json.dumps(data, indent=4))
```

```python
# Select value and store as variable
selected_currency = data["currency"]
selected_price = data["price"]

selected_hometype = data["homeType"]

selected_streetaddress = data["streetAddress"]
selected_city = data["city"]
selected_zipcode = data["zipcode"]
selected_country = data["country"]

# Display the value for the property
print(f"The price of the {selected_hometype} is {selected_currency} {selected_price}")
print(f"Address: {selected_streetaddress}, {selected_city}, {selected_country}, {selected_zipcode}")
```
